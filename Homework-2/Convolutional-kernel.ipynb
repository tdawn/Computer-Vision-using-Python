```python
import cv2
import numpy as np
from pylab import *

image = cv2.imread('234_t.jpg', cv2.IMREAD_GRAYSCALE).astype(float) / 255.0

#Identity
kernel1 = np.array([[0, 0, 0],
                   [0, 1, 0],
                   [0, 0, 0]])
filtered1 = cv2.filter2D(src=image, kernel=kernel1, ddepth=-1)

fig, axs = subplots(1,2, figsize=(6, 6))
axs[0].imshow(image*255)
axs[1].imshow(filtered1*255)
```

```python
#Edge Detection 1
kernel2 = np.array([[1, 0, -1],
                   [0, 0, 0],
                   [-1, 0, 1]])
filtered2 = cv2.filter2D(src=image, kernel=kernel2, ddepth=-1)

#Edge Detection 2
kernel3 = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])
filtered3 = cv2.filter2D(src=image, kernel=kernel3, ddepth=-1)

#Edge Detection 3
kernel4 = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
filtered4 = cv2.filter2D(src=image, kernel=kernel4, ddepth=-1)

fig, axs = subplots(1,4, figsize=(16, 16))
axs[0].imshow(image*255)
axs[1].imshow(filtered2*255)
axs[2].imshow(filtered3*255)
axs[3].imshow(filtered4*255)
```

```python
#Sharpen
kernel5 = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
filtered5 = cv2.filter2D(src=image, kernel=kernel5, ddepth=-1)

fig, axs = subplots(1,2, figsize=(6, 6))
axs[0].imshow(image*255)
axs[1].imshow(filtered5*255)
```

```python
#Box Blur
kernel6 = np.array([[1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])
filtered6 = cv2.filter2D(src=image, kernel=kernel6, ddepth=-1)

#Gaussian Blur 3x3
kernel7 = np.array([[1/16, 1/8, 1/16],
                   [1/8, 1/4, 1/8],
                   [1/16, 1/8, 1/16]])
filtered7 = cv2.filter2D(src=image, kernel=kernel7, ddepth=-1)

#Gaussian Blur 5x5
kernel8 = np.array([[1/256, 4/256, 6/256, 4/256, 1/256],
                   [4/256, 16/256, 24/256, 16/256, 4/256],
                   [6/256, 24/256, 36/256, 24/256, 6/256],
                   [4/256, 16/256, 24/256, 16/256, 4/256],
                   [1/256, 4/256, 6/256, 4/256, 1/256]])
filtered8 = cv2.filter2D(src=image, kernel=kernel8, ddepth=-1)

fig, axs = subplots(1,4, figsize=(16, 16))
axs[0].imshow(image*255)
axs[1].imshow(filtered6*255)
axs[2].imshow(filtered7*255)
axs[3].imshow(filtered8*255)
```
